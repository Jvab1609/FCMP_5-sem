A única forma de fazermos isso funcionar foi voltando para a versão 3.9 do Python e reinstalando as bibliotecas.
Outro adendo é que o modelo .pt deve ser colocado dentro da pasta yolov5, quando ela for clonada

```python
import torch
import cv2
import pathlib
import sys

# Como o modelo foi compilado no Colab usando o Linux, precisa fazer isso para "converter", senão dá erro
if sys.platform == "win32":
    _orig_posixpath = pathlib.PosixPath

    def _patched_posixpath(*args, **kwargs):
        return pathlib.WindowsPath(*args, **kwargs)

    pathlib.PosixPath = _patched_posixpath

# Carregando o modelo

model = torch.hub.load('ultralytics/yolov5', 'custom', path='modelo-800img-50epochs.pt', source='github') 

# Restaurando o posixpath original
if sys.platform == "win32":
    pathlib.PosixPath = _orig_posixpath

# Começa a captura de vídeo
cap = cv2.VideoCapture(1)

if not cap.isOpened():
    print("Erro ao abrir a webcam")
    exit()

while True:
    ret, frame = cap.read()
    if not ret:
        print("Erro ao tentar pegar o quadro")
        break

    # Faz a predição
    results = model(frame)

    # Renderiza os resultados da predição
    annotated_frame = results.render()[0]

    # Mostra a janelinha
    cv2.imshow('YOLOv5 - Detecção', annotated_frame)

    # Para sair, tem que apertar Q. Senão, o kernel morre kkkk
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
